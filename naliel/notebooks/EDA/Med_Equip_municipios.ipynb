```python
def line_equip_munic(data):
    data.rename(columns={'CÃ³digo':'codigo'}, inplace=True)
    data = data.set_index('codigo')
    data_fim = data.drop(data.index[data.shape[0]-1])
    df_out = data_fim.stack()
    df_out.index = df_out.index.map('{0[1]}_{0[0]}'.format)
    return (df_out.to_frame().T)
```

```python
#Run in files
#Para rodar vc deeve Tirar todos os zips e deixar todos os arquivos dentro da pasta equipamentos 

def states_city_equip():
    # List All Folders 
    path_folder = next(os.walk('../data/equipamentos/'))[1]
    df_Fim_todos = pd.DataFrame()
    for val_state in range(0,len(path_folder)-1):
#     for val_state in range(1,1):
        path = '../data/equipamentos/' + path_folder[val_state]
        listOfFiles = os.listdir(path)
        name_cities = [ os.path.splitext(x)[0] for x in listOfFiles]; 
        
#         print('Inicializar estado:', val_state)
        for val_city in range(0,len(listOfFiles)-1):
        # for val_city in range(0,1):
            filename = path + '/'+listOfFiles[val_city]
            filename
            data = pd.read_csv(filename, sep=',', skiprows=0, encoding='latin1')

            data_aux = line_equip_munic(data)
            
            #Inserir Municipios e Estado
            cidades = pd.Series(name_cities[val_city])
            estados = pd.Series(path_folder[val_state])
            data_aux['Municipio'] =  cidades
            data_aux['Estado'] =  estados

            df_Fim_todos = pd.concat([df_Fim_todos, data_aux], axis=0)
#         print('Finalizar estado:', val_state)

    mid = df_Fim_todos['Municipio']
    df_Fim_todos.drop(labels=['Municipio'], axis=1,inplace = True)
    df_Fim_todos.insert(0, 'Municipio', mid)
    df_Fim_todos   

    mid = df_Fim_todos['Estado']
    df_Fim_todos.drop(labels=['Estado'], axis=1,inplace = True)
    df_Fim_todos.insert(0, 'Estado', mid)
    
    display(df_Fim_todos)
    
    return(df_Fim_todos)
```

```python
brasil = states_city_equip()
```
