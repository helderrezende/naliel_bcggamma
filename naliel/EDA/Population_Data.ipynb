```python
def popula_open(path):
    pop_num = pd.read_csv('../data/dados_pop_invt_saude/'+ path, sep=';', skiprows=3, encoding='latin1')
    return pop_num
```

```python
def init_cities(data, key):
    pop_num = data[['Munic-BR','População']]

    aux_colum = pop_num["Munic-BR"].str.split(" ", n = 1, expand = True)
    pop_num["Munic-BR"]= aux_colum[1] 
    pop_num.insert(0, "Val_Municipio", aux_colum[0].values, True) 

    pop_num.insert(0, "Estado",  pop_num["Val_Municipio"].str[0:2], True) 
    pop_num["Val_Municipio"] = pop_num["Val_Municipio"].apply(pd.to_numeric, errors='coerce')
    
    pop_num = pop_num.rename({'Munic-BR': 'Município'}, axis=1)
    pop_num = pop_num.rename({'População': 'Pop_'+str(key)}, axis=1)
    
    return pop_num
```

```python
def all_states(data, key):
    pop_num = data[['Munic-BR','População']]

    aux_colum = pop_num["Munic-BR"].str.split(" ", n = 1, expand = True)
    pop_num["Munic-BR"]= aux_colum[1] 
    pop_num.insert(0, "Val_Municipio", aux_colum[0].values, True) 

    pop_num.insert(0, "Estado",  pop_num["Val_Municipio"].str[0:2], True) 
    pop_num["Val_Municipio"] = pop_num["Val_Municipio"].apply(pd.to_numeric, errors='coerce')
    
    pop_num = pop_num.rename({'Munic-BR': 'Município'}, axis=1)
    pop_num = pop_num.rename({'População': 'Pop_'+str(key)}, axis=1)
    
    #Add collum for state population    
    final_pop =  pop_num.copy()
    final_pop['pop_total_state'] = 0 

    #Add value to column
    UF = 11,12,13,14,15,16,17,21,22,23,24,25,26,27,28,29,31,32,33,35,41,42,43,50,51,52,53
    for uf in UF:
        selec_state =  pop_num.copy()
        selec_state = selec_state[selec_state['Estado']== str(uf)]
        final_pop.loc[final_pop['Estado'] == str(uf), 'pop_total_state'] = selec_state['Pop_'+str(key)].sum() 
    
    #### All statesdisplay(final_pop)
    
    return final_pop

    
```

### All City

```python
dict_pop = {'2014':'A151457177_40_180_122.csv', 
            '2015':'A151417177_40_180_122.csv', 
            '2016':'A153419177_40_180_122.csv', 
            '2017':'A151005177_40_180_122.csv', 
            '2018': 'A150924177_40_180_122.csv'}
```

```python
v=0
for key, value in dict_pop.items():
    data = popula_open(dict_pop[key])
    pop_num_aux = init_cities(data,key)
    
    pop_final = pop_num_aux
    pop_final
```

### All States

```python
p_0 = all_states(popula_open(dict_pop['2014']),'2014')
p_1 = all_states(popula_open(dict_pop['2015']),'2015')
p_2 = all_states(popula_open(dict_pop['2016']),'2016')
p_3 = all_states(popula_open(dict_pop['2017']),'2017')
p_4 = all_states(popula_open(dict_pop['2018']),'2018')

p_0 = p_0.to_csv(r'../data/dados_pop_invt_saude/pop_2014.csv')
p_1 = p_1.to_csv(r'../data/dados_pop_invt_saude/pop_2015.csv')
p_2 = p_2.to_csv(r'../data/dados_pop_invt_saude/pop_2016.csv')
p_3 = p_3.to_csv(r'../data/dados_pop_invt_saude/pop_2017.csv')
p_4 = p_4.to_csv(r'../data/dados_pop_invt_saude/pop_2018.csv')

```
